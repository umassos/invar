```python
import matplotlib.pyplot as plt
import networkx as nx
import numpy as np
```

```python
G = nx.DiGraph()
G.add_node(0)
for i in range(1, 9):
    G.add_node(i)
    G.add_edge(i, 0)
```

```python
data_centers = [0]
user_locations = range(1, 9)
```

```python
pos = {0: np.array([0, 0])}
for i in range(0, 8):
    pos[i + 1] = np.array([np.cos(2 * np.pi * i / 8), np.sin(2 * np.pi * i / 8)])
```

```python
capacity = {0: 18}
```

```python
plt.figure(figsize=(5, 5))
nx.draw_networkx_nodes(G, pos, nodelist=data_centers, node_color="red", node_shape="s", label="data center")
nx.draw_networkx_nodes(G, pos, nodelist=user_locations, node_color="green", label="user_location")
nx.draw_networkx_edges(G, pos, width=5)
nx.draw_networkx_labels(G, pos, capacity)
plt.legend()
plt.savefig("ring_cloud.png")
```
